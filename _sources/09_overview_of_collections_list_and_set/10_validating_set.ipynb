## Validating set

Here are some of the operations that can be performed to validate sets using Python as programming language.

```python
%%HTML
<iframe width="560" height="315" src="https://www.youtube.com/embed/QIYmaH2tiew?rel=0&amp;controls=1&amp;showinfo=0" frameborder="0" allowfullscreen></iframe>
```

* Checking if an element exists (using in operator).
* `issubset` - checking if first set is subset of second set
* `issuperset` - checking if first set is superset of second set
* `isdisjoint` - check if 2 sets have common elements

```python
s = {1, 2, 3, 3, 4, 4, 4, 5}
```

```python
1 in s
```

```python
s1 = {1, 2, 3}
```

```python
s2 = {1, 2, 3, 4, 5}
```

```python
s1.issubset?
```

```python
s1.issubset(s2)
```

```python
s1.issuperset(s2)
```

```python
s2.issuperset(s1)
```

```python
s1.issuperset(s2)
```

```python
s1.isdisjoint?
```

```python
s1 = {1, 2, 3, 4}
```

```python
s2 = {3, 4, 5, 6, 7}
```

```python
s1.isdisjoint(s2)
```
