* Get sum of integers for a given range.

```python
lb = 5
ub = 10

total = 0
for i in range(lb, ub+1):
    total += i
    
print(i)
# Complexity: o(n)
```

```python
lb1 = lb - 1
res1 = lb1 * (lb1 + 1) / 2
print(int(res1))
```

```python
res2 = ub * (ub + 1) / 2
print(int(res2))
```

```python
print(int(res2 - res1))
# Complexity: o(1)
```

```python
# o(1) is much better solution when compared to o(n)
```
